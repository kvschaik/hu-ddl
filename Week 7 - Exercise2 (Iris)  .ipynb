```python
from sklearn.cluster import KMeans #The k-means algorithm
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt



df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.scatterplot(data = df, x = 'PetalLengthCm', y = 'SepalLengthCm', hue = 'Species')
```

```python
#I see two clusters
```

```python
km = KMeans(n_clusters=3)
X = df[['SepalLengthCm', 'PetalLengthCm']] 
```

```python
km = km.fit(X)

df['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
df.head()
```

```python
sns.scatterplot(data = df, x = 'PetalLengthCm', y = 'SepalLengthCm', hue = 'cluster')
```
