```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt


df_movies = pd.read_csv('movies.csv')

df_ratings = pd.read_csv('ratings.csv')
```

```python
df_movies.head()
```

```python
df = pd.merge(df_movies, df_ratings, on = 'movieId')
```

```python
df.head()
```

```python
pivot_table = pd.pivot_table(data=df, index='userId',  columns='title', values='rating')
```

```python
pivot_table.head()
```

```python

pivot_table = pivot_table.reindex(pivot_table.count().sort_values(ascending=False).index, axis =1)
```

```python
pivot_table

pivot_table = pivot_table.fillna(0)
```

```python



pivot_table.corrwith(pivot_table['Toy Story (1995)'], axis=0, drop=False).sort_values(ascending=False).head(10)
```

```python
pivot_table.corrwith(pivot_table['Shawshank Redemption, The (1994)'], axis=0, drop=False).sort_values(ascending=False).head(10)
```
